## Simulation 결과를 활용하여 각 제품 별 초기값 계산

```python
##### 참치액 #####

# 총 판매량
total = 416_000

# 대분류 비율
ratio_sun = 37
ratio_jin = 56
ratio_premium = 7
ratio_sum = ratio_sun + ratio_jin + ratio_premium

# 대분류별 판매량
sun_total = total * ratio_sun / ratio_sum
jin_total = total * ratio_jin / ratio_sum
premium_total = total * ratio_premium / ratio_sum

# 내부 비율
sun_ratio_500 = 3.1
sun_ratio_900 = 1
jin_ratio_500 = 34
jin_ratio_900 = 22
premium_ratio_500 = 6
premium_ratio_900 = 1

# 순: 500g, 900g
sun_500 = sun_total * sun_ratio_500 / (sun_ratio_500 + sun_ratio_900)
sun_900 = sun_total * sun_ratio_900 / (sun_ratio_500 + sun_ratio_900)

# 진: 500g, 900g
jin_500 = jin_total * jin_ratio_500 / (jin_ratio_500 + jin_ratio_900)
jin_900 = jin_total * jin_ratio_900 / (jin_ratio_500 + jin_ratio_900)

# 프리미엄: 500g, 900g
premium_500 = premium_total * premium_ratio_500 / (premium_ratio_500 + premium_ratio_900)
premium_900 = premium_total * premium_ratio_900 / (premium_ratio_500 + premium_ratio_900)

# 결과 출력 (반올림)
print("순:")
print(f"  500g: {round(sun_500)}")
print(f"  900g: {round(sun_900)}")
print("진:")
print(f"  500g: {round(jin_500)}")
print(f"  900g: {round(jin_900)}")
print("프리미엄:")
print(f"  500g: {round(premium_500)}")
print(f"  900g: {round(premium_900)}")
print("\n")


#### 동원 맛참 참치캔 ####

total = 1_250_000

# 비율 설정
ratio_goso = 1.9
ratio_maekom = 1.0

# 고소 & 매콤 개수
goso_total = total * ratio_goso / (ratio_goso + ratio_maekom)
maekom_total = total * ratio_maekom / (ratio_goso + ratio_maekom)

# 내부 비율
goso_ratio_135 = 0.9
goso_ratio_90 = 1.0
maekom_ratio_135 = 0.5
maekom_ratio_90 = 1.0

# 고소 135g, 90g 개수
goso_135 = goso_total * goso_ratio_135 / (goso_ratio_135 + goso_ratio_90)
goso_90 = goso_total * goso_ratio_90 / (goso_ratio_135 + goso_ratio_90)

# 매콤 135g, 90g 개수
maekom_135 = maekom_total * maekom_ratio_135 / (maekom_ratio_135 + maekom_ratio_90)
maekom_90 = maekom_total * maekom_ratio_90 / (maekom_ratio_135 + maekom_ratio_90)

# 결과 출력
print(f"고소 총합: {goso_total:.0f} 캔")
print(f" - 135g: {goso_135:.0f} 캔")
print(f" - 90g:  {goso_90:.0f} 캔")
print(f"매콤 총합: {maekom_total:.0f} 캔")
print(f" - 135g: {maekom_135:.0f} 캔")
print(f" - 90g:  {maekom_90:.0f} 캔")

print("\n")
##### 라떼 #####
# 150만개에서 시작 -> 2달 후 100만개로 감소
##### 카페 라떼 #####
cafe_latte = 1500000 * (2.6 / 3.6)
print("Cafe Latte :", int(cafe_latte))
##### 바닐라 라떼 #####
vanila_latte = 1500000 * (1 / 3.6)
print("Vanilla Latte :", int(vanila_latte))

print("\n")
# 오믈레햄
# 월 판매액 15만개
###### 200g ######
ham_200 = 150000 * (2.5 / 3.5)
print("오믈레햄 200g :", int(ham_200))

###### 340g ######
ham_340 = 150000 * (1 / 3.5)
print("오믈레햄 340g :", int(ham_340))
```

## 뉴스 정보에 따라 라떼 제품 군 마지막 2달은 2/3 곱하기

```python
from datetime import date

# 1) 대상 제품
latte_products = [
    "소화가 잘되는 우유로 만든 바닐라라떼 250mL",
    "소화가 잘되는 우유로 만든 카페라떼 250mL",
]

# 2) 절대 달 매핑(2024-07 ~ 2025-06)
start_y, start_m = 2024, 7
abs_months = []
y, m = start_y, start_m
for _ in range(12):
    abs_months.append((y, m))
    m += 1
    if m == 13:
        y += 1
        m = 1

idx_by_ym = {ym: i+1 for i, ym in enumerate(abs_months)}  # {(2024,7):1, ..., (2025,6):12}

# 3) 2025-05, 2025-06에 해당하는 컬럼 계산
targets_ym = [(2025, 5), (2025, 6)]
target_cols = [f"months_since_launch_{idx_by_ym[ym]}"
               for ym in targets_ym if ym in idx_by_ym]

# 4) 라떼 제품 행에만 해당 달 컬럼에 2/3 곱해서 정수화
mask = df["product_name"].isin(latte_products)
if target_cols: 
    df.loc[mask, target_cols] = (df.loc[mask, target_cols] * (2/3)).astype(int)

df[month_cols] = df[month_cols].astype(int)
```

### 제품 출시일을 고려하여 출시 전까지는 0으로 보정 / 출시 달은 실제 판매일만큼의 비율만을 곱해준다.

```python
from datetime import date
import calendar

# 0) 절대 달 매핑: months_since_launch_1 = 2024-07, ..., _12 = 2025-06
abs_months = []
y, m = 2024, 7
for _ in range(12):
    abs_months.append((y, m))
    m += 1
    if m == 13:
        y += 1
        m = 1

col_by_idx = {i+1: f"months_since_launch_{i+1}" for i in range(12)}
idx_by_ym  = {ym: i+1 for i, ym in enumerate(abs_months)}  # {(2024,7):1, ...}

# 1) 출시일 정보
launch_dates = {
    "덴마크 하이그릭요거트 400g": date(2025, 2, 7),
    "리챔 오믈레햄 200g": date(2025, 5, 22),
    "리챔 오믈레햄 340g": date(2025, 5, 22),
    "소화가 잘되는 우유로 만든 바닐라라떼 250mL": date(2025, 2, 21),
    "소화가 잘되는 우유로 만든 카페라떼 250mL": date(2025, 2, 21),
}

# 2) 제품별로 출시월 이전=0, 출시월=부분월 가중, 이후=그대로
for pname, ldt in launch_dates.items():
    mask = (df["product_name"] == pname)
    if not mask.any():
        continue

    launch_ym = (ldt.year, ldt.month)
    launch_idx = idx_by_ym.get(launch_ym, None)

    # 윈도우(2024-07~2025-06)에 출시월이 없다면 케이스 처리
    if launch_idx is None:
        # 출시가 윈도우 이후라면 전부 0, 이전이라면 전부 그대로
        if launch_ym > abs_months[-1]:
            df.loc[mask, list(col_by_idx.values())] = 0
        # 출시가 윈도우 이전이면 할 일 없음(이미 전월이 없음)
        continue

    # (a) 출시월 이전 달들은 0으로
    pre_cols = [col_by_idx[i] for i in range(1, launch_idx)]
    if pre_cols:
        df.loc[mask, pre_cols] = 0

    # (b) 출시월 부분월 가중치
    days_in_month = calendar.monthrange(ldt.year, ldt.month)[1]
    sell_days = days_in_month - ldt.day + 1
    factor = sell_days / days_in_month

    launch_col = col_by_idx[launch_idx]
    df.loc[mask, launch_col] = (df.loc[mask, launch_col] * factor).astype(int)

# 전체 월 정수화 안전 처리
month_cols = [f"months_since_launch_{i}" for i in range(1, 13)]
df[month_cols] = df[month_cols].round().astype(int)
```

## 제품 별 성장 Trend 추가

```python
import numpy as np

# 1) 2% 월별 성장 계수 (12개월)
growth = np.array([1.04**(i) for i in range(1, 13)])  # [1.00, 1.04, ..., 1.04**11]

# 2) 대상 제품 마스크
mask = df["product_name"].isin(["동원참치액 진 500g"])

# 3) 월별 컬럼에 성장 계수 적용 (정수 반올림)
df.loc[mask, month_cols] = (
    df.loc[mask, month_cols]
      .astype(float)
      .multiply(growth, axis=1)
      .round(0)
      .astype(int)
)
```
